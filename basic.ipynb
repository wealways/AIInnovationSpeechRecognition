```python
# 클래스 (목적: 재사용)
class Operation:
    def __init__(self):
        self.a = 1
    
    def plus(self, b):
        return (self.a + b)
    
    def minus(self, b):
        self.c = self.a - b
        return (self.a - b)
    
    

op = Operation()
print(op.plus(2))
print(op.minus(3))
print(op.c)
```

```python
from PIL import Image

cat_jpg = Image.open('./img/cat.jpg')
cat_jpg.size
```

```python
cat_jpg.resize((400, 400))
```

```python
from matplotlib import pyplot as plt


cat_jpg = np.array(cat_jpg)

plt.imshow(cat_jpg)
plt.axis("off")
plt.show()
```

```python
print(type(cat_jpg))
print(cat_jpg.ndim)
print(cat_jpg.shape)
print(cat_jpg)
```

```python
img_rgb = cat_jpg

plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray)
plt.axis("off")
plt.title("RGB 이미지")

plt.subplot(222)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.gray)
plt.axis("off")
plt.title("R 채널")

plt.subplot(223)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.gray)
plt.axis("off")
plt.title("G 채널")

plt.subplot(224)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.gray)
plt.axis("off")
plt.title("B 채널")

plt.show()
```
